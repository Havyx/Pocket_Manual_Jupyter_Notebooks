The class label indicates if the person has a coronary heart disease: negative (0) or positive (1). 

Attributes description: 
- sbp: systolic blood pressure 
- tobacco: cumulative tobacco (kg) 
- ldl: low densiity lipoprotein cholesterol 
- adiposity 
- famhist: family history of heart disease (Present, Absent) 
- typea: type-A behavior 
- obesity 
- alcohol: current alcohol consumption 
- age: age at onset

```python
df = pd.read_csv('00_database/dados.csv')
df.head()
```

### Balanceando os dados por undersample

```python
df.chd.sum() / df.chd.shape[0]
```

```python
df_zeros = df.loc[df.chd==0]
df_zeros = df_zeros.iloc[:(int(df_zeros.shape[0]/2))]
df_zeros
```

```python
df_ones = df.loc[df.chd==1]
df_ones
```

### DUMMIES DA VARIAVEL CATEGORICA 

```python
df = pd.concat((df_ones, df_zeros))
df
```

### Dados Balanceados

```python
df.chd.sum() / df.chd.shape[0]
```

```python
df = pd.concat((df, pd.get_dummies(df.famhist)), axis=1)
df.drop('famhist', inplace=True, axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
colunas = ['sbp', 'tobacco', 'ldl', 'adiposity', 'typea', 'obesity', 'alcohol', 'age']
```

### Normalizando os dados

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
df_s = df.copy()
df_s[colunas] = scaler.fit_transform(df[colunas])
df_s
```

### Aplicando regrecao logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
y = df_s['chd']
X = df_s.drop('chd', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_test.to_csv('novos_dados.csv', index=False)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
predito = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predito)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predito))
```

```python
import numpy as np
import pandas as pd
import pickle

class SavioLogisticModel():
      
        def __init__(self, model_file):
            with open('model.pkl','rb') as model_file:
                self.reg = pickle.load(model_file)
                self.data = None
                
        def load_new_data(self, data_file):
            df = pd.read_csv(data_file,delimiter=',')
            df = df.sample(10)
            self.data = df.copy()
            self.preprocessed_data = df.copy()
            
        def predicted_probability(self):
            if (self.data is not None):  
                pred = self.reg.predict_proba(self.data)
                return pred
            
        def predicted_output_category(self):
            if (self.data is not None):
                pred_outputs = self.reg.predict(self.data)
                return pred_outputs
            
        def predicted_outputs(self):
            if (self.data is not None):
                self.preprocessed_data['Probabilidade'] = 0
                self.preprocessed_data['Predito'] = 0
                self.preprocessed_data['Probabilidade'] = self.reg.predict_proba(self.data)[:,1]
                self.preprocessed_data['Predito'] = self.reg.predict(self.data)
                return self.preprocessed_data
```

```python
model = SavioLogisticModel('model.pkl')
model.load_new_data('novos_dados.csv')
model.predicted_probability()[:,1]
model.predicted_output_category()
model.predicted_outputs()
```
