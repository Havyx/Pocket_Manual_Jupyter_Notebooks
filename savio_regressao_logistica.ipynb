The class label indicates if the person has a coronary heart disease: negative (0) or positive (1). 

Attributes description: 
- sbp: systolic blood pressure 
- tobacco: cumulative tobacco (kg) 
- ldl: low densiity lipoprotein cholesterol 
- adiposity 
- famhist: family history of heart disease (Present, Absent) 
- typea: type-A behavior 
- obesity 
- alcohol: current alcohol consumption 
- age: age at onset

```python
df = pd.read_csv('dados.csv')
df.head()
```

### Balanceando os dados por undersample

```python
df.chd.sum() / df.chd.shape[0]
```

```python
df_zeros = df.loc[df.chd==0]
df_zeros = df_zeros.iloc[:(int(df_zeros.shape[0]/2))]
df_zeros
```

```python
df_ones = df.loc[df.chd==1]
df_ones
```

### DUMMIES DA VARIAVEL CATEGORICA 

```python
df = pd.concat((df_ones, df_zeros))
df
```

### Dados Balanceados

```python
df.chd.sum() / df.chd.shape[0]
```

```python
df = pd.concat((df, pd.get_dummies(df.famhist)), axis=1)
df.drop('famhist', inplace=True, axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
colunas = ['sbp', 'tobacco', 'ldl', 'adiposity', 'typea', 'obesity', 'alcohol', 'age']
```

### Normalizando os dados

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
df_s = df.copy(* Lembrar de treinar o algorimo from scratch quando estudar js)
df_s[colunas] = scaler.fit_transform(df[colunas])
df_s
```

### Aplicando regrecao logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
y = df_s['chd']
X = df_s.drop('chd', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
predito = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predito)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predito))
```
